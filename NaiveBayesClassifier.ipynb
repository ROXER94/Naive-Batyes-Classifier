## Naive Bayes text classifier

```python
ptexts=[open('ps2data/p{}.txt'.format(i)).read() for i in range(20)]
stexts=[open('ps2data/s{}.txt'.format(i)).read() for i in range(20)]
ttexts=[open('ps2data/t{}.txt'.format(i)).read() for i in range(10)]
from urllib2 import urlopen
from zipfile import ZipFile
from StringIO import StringIO

ps2texts=urlopen('https://courses.cit.cornell.edu/info2950_2014fa/resources/ps2data.zip')
with ZipFile(StringIO(ps2texts.read())) as ziptexts:
    ptexts=[ziptexts.read('ps2data/p{}.txt'.format(i)) for i in range(20)]
    stexts=[ziptexts.read('ps2data/s{}.txt'.format(i)) for i in range(20)]
    ttexts=[ziptexts.read('ps2data/t{}.txt'.format(i)) for i in range(10)]
ps2texts.close()
import re
re.findall("[a-z']+",ptexts[0].lower())
from collections import Counter
Counter(set(re.findall("[a-z']+",ptexts[0].lower())))
np = sum([Counter(set(re.findall("[a-z']+",txt.lower()))) for txt in ptexts])
ns = sum([Counter(set(re.findall("[a-z']+",txt.lower()))) for txt in stexts])
#np.most_common(10),ns.most_common(10)
#np.most_common()[62:72]

ptexts=[open('ps2data/p{}.txt'.format(i)).read() for i in range(20)]
stexts=[open('ps2data/s{}.txt'.format(i)).read() for i in range(20)]
ttexts=[open('ps2data/t{}.txt'.format(i)).read() for i in range(10)]
def bayes_classifier(txt):
    word_counts = Counter(re.findall("[a-z']+",txt.lower()))
    keywords=[w[0] for w in word_counts.most_common(30)]
    varns = 1.
    varnp = 1.
    for word in keywords:
        if word not in np: np[word] = .1
        if word not in ns: ns[word] = .1
        varnp = varnp*np[word]/20.
        varns = varns*ns[word]/20.
	prob = (.5*varnp)/(.5*varnp+.5*varns)
    if prob >= .5:
    	return 'politics',prob
    else:
        return 'sports',prob
for i in range(10):
    print 't{}.txt:'.format(i),bayes_classifier(ttexts[i])
```
